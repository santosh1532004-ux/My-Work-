**QUESTION 1- Explain the diffrence between Relative ,Absolute and Mixed cell Refrencing **

ANS- **Relative cell refrencing** -Cell refrence changes automatically when the formula is copied .

    * Why it is used* - when the same calculation is needed for multiple rows .
                   -Most common and default refrencing type .
                    REAL USE - Calculating the total marks for each student row wise .
     **Absolute cell refrencing **- Cell refrence remains fixed ,even  when the formula is copied .

                    *Why it is used* - When one fixed value like tax rate  discount rate is used everywhere .



      **MIXED cell refrencing **-One part (row or column)is fixed, the other changes

    WHY it is used -Useful in tables, matrices, multiplication grids .
